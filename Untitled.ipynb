```python
import pandas as pd
import os

# Proje dizinini kontrol edelim
BASE_DIR = os.getcwd()
print("Notebook dizini:", BASE_DIR)

# data_raw klasörüne giden yol
DATA_RAW_DIR = os.path.join(os.path.dirname(BASE_DIR), "data_raw")
print("Veri dizini:", DATA_RAW_DIR)

# CSV dosyalarını listele
csv_files = [f for f in os.listdir(DATA_RAW_DIR) if f.endswith(".csv")]
print("CSV sayısı:", len(csv_files))

csv_files[:5]  # ilk 5 dosyayı göster
```

```python
df_list = []

for file in csv_files:
    file_path = os.path.join(DATA_RAW_DIR, file)
    temp_df = pd.read_csv(file_path)
    df_list.append(temp_df)

df = pd.concat(df_list, ignore_index=True)

print("Birleştirilmiş veri boyutu:", df.shape)
df.head()
```

```python
# ft sütununu datetime'a çevir
df["ft"] = pd.to_datetime(df["ft"], errors="coerce")

# Zamana göre sırala
df = df.sort_values("ft").reset_index(drop=True)

df[["ft"]].head()
```

```python
selected_columns = [
    "ft",
    "ANALOGS_BATTERY_VOLTAGE",
    "ANALOGS_BATTERY_1_CURRENT",
    "ANALOGS_BUS_TEMP",
    "RADIO_SDR_TEMP",
    "GPS_MSG_TRACKED_SATELLITES"
]

df_sel = df[selected_columns].copy()
df_sel.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,4))
plt.plot(df_sel["ft"], df_sel["ANALOGS_BATTERY_VOLTAGE"])
plt.title("Battery Voltage Over Time")
plt.xlabel("Time")
plt.ylabel("Voltage")
plt.show()
```

```python
from scipy.stats import zscore

df_sel["battery_voltage_z"] = zscore(df_sel["ANALOGS_BATTERY_VOLTAGE"].fillna(method="ffill"))

# |z| > 3 olanları anomali kabul edelim
df_sel["battery_voltage_anomaly"] = df_sel["battery_voltage_z"].abs() > 3

df_sel[df_sel["battery_voltage_anomaly"]].head()
```

```python
from scipy.stats import zscore

anomaly_features = [
    "ANALOGS_BATTERY_VOLTAGE",
    "ANALOGS_BATTERY_1_CURRENT",
    "ANALOGS_BUS_TEMP",
    "RADIO_SDR_TEMP",
    "GPS_MSG_TRACKED_SATELLITES"
]

for col in anomaly_features:
    df_sel[col + "_z"] = zscore(df_sel[col].fillna(method="ffill"))
    df_sel[col + "_anomaly"] = df_sel[col + "_z"].abs() > 3

df_sel[[c for c in df_sel.columns if "anomaly" in c]].sum()
```

```python
events = []

for col in anomaly_features:
    anomalies = df_sel[df_sel[col + "_anomaly"]]
    for _, row in anomalies.iterrows():
        events.append({
            "time": row["ft"],
            "parameter": col,
            "value": row[col],
            "z_score": row[col + "_z"]
        })

events_df = pd.DataFrame(events)
events_df.head()
```

```python
def anomaly_to_text(event):
    return (
        f"At {event['time']}, an anomaly was detected in {event['parameter']}. "
        f"The observed value was {event['value']:.3f}, "
        f"which deviates significantly from normal behavior (z-score = {event['z_score']:.2f}). "
        f"This may indicate a potential issue in the related subsystem."
    )

events_df["llm_input"] = events_df.apply(anomaly_to_text, axis=1)
events_df["llm_input"].head()
```

```python
def build_llm_prompt(event_text):
    prompt = f"""
You are a spacecraft systems engineer.

The following telemetry anomaly was detected:
"{event_text}"

Please provide:
1. Possible technical reasons for this anomaly
2. Potential risk level (Low / Medium / High)
3. Recommended actions or checks

Answer in clear, concise technical language.
"""
    return prompt


# test edelim
sample_prompt = build_llm_prompt(events_df["llm_input"].iloc[0])
print(sample_prompt)
```

```python
def mock_llm_response(prompt):
    response = {
        "possible_technical_reasons": [
            "Temporary battery discharge due to increased power demand",
            "Reduced solar charging efficiency during orbital eclipse",
            "Battery aging or internal resistance increase"
        ],
        "risk_level": "Medium",
        "recommended_actions": [
            "Monitor battery voltage trend over the next orbits",
            "Check power subsystem logs for abnormal load conditions",
            "Verify thermal conditions affecting battery performance"
        ]
    }
    return response


# Az önce oluşturduğun prompt'u kullanalım
llm_result = mock_llm_response(sample_prompt)

llm_result
```

```python
def build_final_record(event, llm_output):
    return {
        "time": event["time"],
        "parameter": event["parameter"],
        "observed_value": round(event["value"], 3),
        "z_score": round(event["z_score"], 2),
        "risk_level": llm_output["risk_level"],
        "possible_technical_reasons": "; ".join(llm_output["possible_technical_reasons"]),
        "recommended_actions": "; ".join(llm_output["recommended_actions"])
    }


# İlk anomaly event'i kullanalım
event = events_df.iloc[0]

final_record = build_final_record(event, llm_result)

final_record
```

```python
final_df = pd.DataFrame([final_record])
final_df
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
from peft import LoraConfig, get_peft_model

model_name = "distilgpt2"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
lora_config = LoraConfig(
    r=8,
    lora_alpha=16,
    target_modules=["c_attn"],
    lora_dropout=0.1,
    bias="none",
    task_type="CAUSAL_LM"
)

model = get_peft_model(model, lora_config)

model.print_trainable_parameters()
```

```python
from datasets import Dataset

train_texts = events_df["llm_input"].head(10).tolist()

def tokenize(batch):
    return tokenizer(
        batch,
        truncation=True,
        padding="max_length",
        max_length=128
    )

dataset = Dataset.from_dict({"text": train_texts})
tokenized_dataset = dataset.map(lambda x: tokenize(x["text"]))
```

```python
def tokenize(batch):
    return tokenizer(
        batch["text"],
        truncation=True,
        padding="max_length",
        max_length=128
    )

tokenized_dataset = dataset.map(tokenize, batched=True)
```

```python
tokenizer.pad_token = tokenizer.eos_token
```

```python
tokenized_dataset[0].keys()
```

```python
from transformers import Trainer, TrainingArguments

training_args = TrainingArguments(
    output_dir="./lora_demo",
    per_device_train_batch_size=2,
    num_train_epochs=1,   # 🔴 bilinçli karar
    logging_steps=1,
    save_strategy="no",
    report_to="none"
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset
)

trainer.train()
```

```python
from transformers import DataCollatorForLanguageModeling

data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False   # GPT tipi model olduğu için
)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    data_collator=data_collator
)
```

```python
def alert_operator(event_row):
    print("🚨 ANOMALY DETECTED 🚨")
    print(f"Time       : {event_row['time']}")
    print(f"Parameter  : {event_row['parameter']}")
    print(f"Value      : {event_row['observed_value']}")
    print(f"Z-score    : {event_row['z_score']}")
    print(f"Risk Level : {event_row['risk_level']}")
    print("\n🛠 Possible Reasons:")
    print(event_row['possible_technical_reasons'])
    print("\n✅ Recommended Actions:")
    print(event_row['recommended_actions'])
    print("-" * 50)
```

```python
anomaly_counter = 0
```

```python
def alert_operator(event_row):
    global anomaly_counter
    anomaly_counter += 1

    print("\n🚨🚨🚨 ANOMALY DETECTED 🚨🚨🚨")
    print(f"Alarm No   : {anomaly_counter}")
    print(f"Time       : {event_row['time']}")
    print(f"Parameter  : {event_row['parameter']}")
    print(f"Value      : {event_row['observed_value']}")
    print(f"Z-score    : {event_row['z_score']}")
    print(f"Risk Level : {event_row['risk_level']}")

    print("\n🧠 SYSTEM INTERPRETATION:")
    print(event_row['possible_technical_reasons'])

    print("\n✅ RECOMMENDED ACTIONS:")
    print(event_row['recommended_actions'])

    print("-" * 60)
```

```python
import time

def real_time_simulation(events_df, delay=0.5):
    print("🛰️ Starting real-time telemetry monitoring...\n")

    for _, event in events_df.iterrows():
        print(f"⏱ Telemetry received at {event['time']}")

        # Anomali varsa alarm ver
        alert_operator(event)

        time.sleep(delay)

    print("\n✅ Monitoring finished.")
    print(f"📊 Total anomalies detected: {anomaly_counter}")
```
