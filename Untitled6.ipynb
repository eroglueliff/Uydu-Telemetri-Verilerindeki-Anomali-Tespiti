```python
import pandas as pd
import numpy as np
import os
import torch
import matplotlib.pyplot as plt
from scipy.stats import zscore
from transformers import AutoModelForCausalLM, AutoTokenizer, TrainingArguments, Trainer, DataCollatorForLanguageModeling
from peft import LoraConfig, get_peft_model, TaskType
from datasets import Dataset

# 1. SISTEM HAZIRLIÄžI
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"System Ready. Device: {device}")

# 2. VERÄ° YÃœKLEME VE Ã–N Ä°ÅžLEME

BASE_DIR = os.getcwd()
DATA_RAW_DIR = os.path.join(os.path.dirname(BASE_DIR), "data_raw")

try:
    csv_files = sorted([f for f in os.listdir(DATA_RAW_DIR) if f.endswith(".csv")])
    df = pd.concat([pd.read_csv(os.path.join(DATA_RAW_DIR, f)) for f in csv_files], ignore_index=True)
    df["ft"] = pd.to_datetime(df["ft"], errors="coerce")
    df = df.sort_values("ft").reset_index(drop=True)
    print(f"Data loaded. Total rows: {len(df)}")
except Exception as e:
    print(f"Error loading data: {e}")

# Anomaliler Ä°Ã§in Kritik Parametre GruplarÄ±
thermal_cols = ["ANALOGS_BUS_TEMP", "RADIO_SDR_TEMP", "ANALOGS_BATTERY1_TEMP", "ANALOGS_PL_TIRS_TEMP1"]
power_cols = ["ANALOGS_BATTERY_VOLTAGE", "ANALOGS_BATTERY_1_CURRENT"]
nav_cols = ["GPS_MSG_TRACKED_SATELLITES"]
env_cols = ["REFS_SUN_ECLIPSE_EARTH_UMBRA_FLAG"]

target_cols = thermal_cols + power_cols + nav_cols
df_sel = df[["ft"] + target_cols + env_cols].copy()

# 3. ANOMALÄ° TESPÄ°TÄ° (Z-Score)
for col in target_cols:
    df_sel[col] = df_sel[col].fillna(method="ffill")
    df_sel[f"{col}_z"] = zscore(df_sel[col])
    df_sel[f"{col}_anomaly"] = df_sel[f"{col}_z"].abs() > 3

anomaly_indices = df_sel[df_sel[[c for c in df_sel.columns if "_anomaly" in c]].any(axis=1)].index
print(f"Total anomalies detected: {len(anomaly_indices)}")

# 4. Model seÃ§imi  ve Tokenizer hazÄ±rlÄ±ÄŸÄ±
model_name = "distilgpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token

model = AutoModelForCausalLM.from_pretrained(model_name).to(device)

# LoRA (Low-Rank Adaptation) AyarlarÄ±
lora_config = LoraConfig(
    r=16, 
    lora_alpha=32,
    target_modules=["c_attn"],
    lora_dropout=0.1, 
    task_type=TaskType.CAUSAL_LM
)
model = get_peft_model(model, lora_config)

# 5. Ä°NGÄ°LÄ°ZCE PROMPT VE EÄžÄ°TÄ°M VERÄ°SÄ° OLUÅžTURMA
def create_structured_prompt(row):
    t_sum = ", ".join([f"{c.split('_')[1]}:{row[c]:.1f}C" for c in thermal_cols])
    env = "ECLIPSE" if row["REFS_SUN_ECLIPSE_EARTH_UMBRA_FLAG"] == "YES" else "SUNLIGHT"
    return f"### Satellite Report\nEnvironment: {env}\nThermal: {t_sum}\nVoltage: {row['ANALOGS_BATTERY_VOLTAGE']:.2f}V\nAnalysis:"

def build_train_sample(row):
    prompt = create_structured_prompt(row)
    if row["ANALOGS_BATTERY_VOLTAGE"] < 12.05 and row["REFS_SUN_ECLIPSE_EARTH_UMBRA_FLAG"] == "YES":
        ans = "Expected voltage drop due to eclipse transition. Risk: Low."
    elif row["RADIO_SDR_TEMP"] > 11.5:
        ans = "Temperature spike detected in SDR unit. Cooling required. Risk: Medium."
    else:
        ans = "Telemetry variance detected. System within operational limits. Risk: Low."
    return f"{prompt} {ans} {tokenizer.eos_token}"

# EÄŸitim seti hazÄ±rlÄ±ÄŸÄ± (Ä°lk 200 anomaliyi kullanÄ±yoruz)
train_texts = [build_train_sample(df_sel.iloc[i]) for i in anomaly_indices[:200]]
train_dataset = Dataset.from_dict({"text": train_texts})
tokenized_dataset = train_dataset.map(
    lambda x: tokenizer(x["text"], truncation=True, padding="max_length", max_length=128), 
    batched=True
)

# 6. EÄžÄ°TÄ°M (FINE-TUNING)
training_args = TrainingArguments(
    output_dir="./satellite_en_v1",
    per_device_train_batch_size=4,
    num_train_epochs=10,
    learning_rate=1e-4,
    logging_steps=5,
    save_strategy="no",
    report_to="none"
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    data_collator=DataCollatorForLanguageModeling(tokenizer, mlm=False)
)

print("Training started...")
trainer.train()

# 7. RAPORLAMA VE ANALÄ°Z FONKSÄ°YONU
def process_anomalies_to_excel():
    results = []
    subset_indices = anomaly_indices[:100] # Ä°lk 100 anomaliyi raporla
    print(f"ðŸ”„ Generating reports for {len(subset_indices)} points...")
    
    model.eval()
    for idx in subset_indices:
        row = df_sel.iloc[idx]
        prompt = create_structured_prompt(row)
        inputs = tokenizer(prompt, return_tensors="pt").to(device)
        
        with torch.no_grad():
            outputs = model.generate(
                **inputs, 
                max_new_tokens=35, 
                repetition_penalty=2.5,
                no_repeat_ngram_size=3,
                temperature=0.3, # Daha az halÃ¼sinasyon iÃ§in dÃ¼ÅŸÃ¼k sÄ±caklÄ±k
                do_sample=True,
                pad_token_id=tokenizer.eos_token_id
            )
        
        result = tokenizer.decode(outputs[0], skip_special_tokens=True)
        
        final_output = result.split("Analysis:")[-1].strip()
        
        # Guardrail (Yedek Mekanizma)
        if len(final_output) < 5 or "_" in final_output:
            if row["ANALOGS_BATTERY_VOLTAGE"] < 12.05:
                final_output = "Nominal voltage drop during eclipse. Risk: Low."
            else:
                final_output = "Anomalous telemetry detected. Monitor system state. Risk: Medium."

        # Zaman damgasÄ±nÄ± temizleyelim Ã§Ã¼nkÃ¼ hata aldÄ±k
        clean_time = row["ft"].replace(tzinfo=None) if hasattr(row["ft"], "tzinfo") else row["ft"]

        results.append({
            "Timestamp": clean_time,
            "Mode": "ECLIPSE" if row["REFS_SUN_ECLIPSE_EARTH_UMBRA_FLAG"] == "YES" else "SUNLIGHT",
            "Voltage": row["ANALOGS_BATTERY_VOLTAGE"],
            "Bus_Temp": row["ANALOGS_BUS_TEMP"],
            "SDR_Temp": row["RADIO_SDR_TEMP"],
            "AI_Analysis": final_output
        })

    report_df = pd.DataFrame(results)
    report_df.to_excel("satellite_technical_report_en.xlsx", index=False)
    print("\nâœ… Final report saved as: satellite_technical_report_en.xlsx")
    
    # Loss GrafiÄŸi
    history = trainer.state.log_history
    loss_values = [log['loss'] for log in history if 'loss' in log]
    plt.figure(figsize=(8, 4))
    plt.plot(loss_values, label='Training Loss')
    plt.title('Model Training Progress (English)')
    plt.xlabel('Steps')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()

# Ã‡ALIÅžTIR
process_anomalies_to_excel()
```
